```python
# popular dwelling involved when buying the houses
# how 'MSSubClass' affect the sale price
ax = df['MSSubClass'].value_counts()\
    .plot(kind='bar')

ax.set_ylabel('Number of Houses', fontsize=12)
ax.set_xlabel('Type of Dwelling', fontsize=12)
ax.set_title("Dwelling involved in the Sale", fontsize=14, fontweight='bold')

plt.show()
```

When it comes to dwelling involved, (20)1-STORY 1946 & NEWER ALL STYLES is the most popular one and the second one is (60)2-STORY 1946 & NEWER

```python
# how MSZoning is in the data
ax = df['MSZoning'].value_counts()\
    .plot(kind='barh')

ax.set_xlabel('Number of Houses', fontsize=12)
ax.set_ylabel('Zone Type', fontsize=12)
ax.set_title("Popular Zones", fontsize=14, fontweight='bold')

plt.show()
```

Popular zone type is RL - Residential Low Density

```python
# how type of dewelling affecting affecting the sale price of the houses
ax = sns.boxplot(df, x = 'MSSubClass', y = 'SalePrice',
                 flierprops={"marker":"o", "markerfacecolor":"red", "markeredgecolor":"black", "markersize":6}) #style outliers

ax.set_xlabel("Dwelling Type(MSSubClass)", fontsize=12)
ax.set_ylabel("Sale Price", fontsize=12)
ax.set_title("Distribution of Sale Price by Dwelling Type", fontsize=14, fontweight='bold')



plt.show()
```

The Popular 20, 60 dwelling types are having most outliers

```python
# How zone type affecting the sale price is there any influence from zone
ax = sns.barplot(df, x = 'MSZoning', y = 'SalePrice', estimator=np.median)

ax.set_ylabel('Sale Price', fontsize=12)
ax.set_xlabel('Zone Type', fontsize=12)
ax.set_title('Popular Zone Types', fontsize=14, fontweight='bold')

plt.show()
```

Popular zone type RL but it only second to the demand because FV (Floating Village Residential) is on the high price

```python
# 'HouseStyle' 'Street' 'OverallCond' 'RoofStyle' 'RoofMatl' 'GarageCars' 'GarageYrBlt' 'SaleCondition'
cols = ['OverallCond', 'OverallQual']

# check relationship strength
sns.heatmap(df[['OverallCond', 'OverallQual', 'SalePrice']].corr(), annot=True, linewidths=0.8)

plt.show()
```

```python
ax = sns.scatterplot(df, x = 'YearBuilt', y = 'SalePrice', alpha=0.5, hue='YearSold',
                     palette='viridis', size='YearSold', sizes=(20, 150))

ax.set_xlabel("Year Built", fontsize=12)
ax.set_ylabel("Sale Price", fontsize=12)
ax.legend(title='Year Sold', fontsize=10)

ax.set_title("Sale Price vs. Year Built (Colored by Year Sold)", fontsize=14, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
